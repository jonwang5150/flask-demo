```python
programs=[
 '新店-自主訓練月會籍',
 '新店-自主訓練季會籍',
 '新店-自主訓練年會籍',
 '新店-單月場租教練吃到飽會籍',
 '新店-單月場租計次會籍',
 '新店-單月團課吃到飽會籍',
 '新店-團體月會籍',
 '新店-團體季會籍',
 '新店-團體年會籍',
 ]
```

```python
df.info()
```

```python
df['電話']=df['電話'].astype(str)
```

```python
df['電話']=df['電話'].apply(lambda x:'0'+x)
```

```python
df['建立時間']=df['建立時間'].apply(pd.to_datetime)
```

```python
df['訂單生效日']=df['訂單生效日'].apply(pd.to_datetime)
```

```python
writer=pd.ExcelWriter('會籍9M.xlsx',engine='xlsxwriter')
filters=[
  (df['建立時間']>'2023-08-01') & (df['建立時間']<'2023-08-31'),
  (df['建立時間']>'2023-09-01') & (df['建立時間']<'2023-09-30'),
  (df['建立時間']>'2023-10-01') & (df['建立時間']<'2023-10-31'),
  (df['建立時間']>'2023-11-01') & (df['建立時間']<'2023-11-30'),
  (df['建立時間']>'2023-11-01') & (df['建立時間']<'2023-11-30'),
  (df['建立時間']>'2023-12-01') & (df['建立時間']<'2023-12-31'),
]
for i,filter in enumerate(filters):
  writer=pd.ExcelWriter(f'會籍{8+i}M.xlsx',engine='xlsxwriter')
  for program in programs:
    df_new=df.groupby('品名').get_group(program)[filter].sort_values('建立時間')
    df_new=df_new.reset_index(drop=True)
    df_new['建立時間']=df_new['建立時間'].dt.date
    df_new['訂單生效日']=df_new['訂單生效日'].dt.date
    df_new.to_excel(writer,sheet_name=program)

  writer.save()
```
